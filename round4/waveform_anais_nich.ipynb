```python
import xarray as xr
import pandas as pd
import seaborn as sns
import numpy as np
import datashader as ds
from datashader import transfer_functions as tf
```

```python
dataset = xr.load_dataset('steinmetz_neuromatch_dataset/steinmetz_2016-12-14_Cori.nc')[['waveform_w', 'trough_to_peak']]
dataset
```

```python
dataset_wf = dataset.sel(waveform_component=1).to_dataframe().reset_index()
dataset_wf
sns.lineplot(data=dataset_wf, x='sample', y='waveform_w', hue='cell')
```

```python
canvas = ds.Canvas(plot_width=800, plot_height=500)
agg = canvas.line(dataset_wf, 'sample', 'waveform_w', agg=ds.count())
plot = tf.shade(agg)
plot = tf.set_background(plot, "white")
plot
```
